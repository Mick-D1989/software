## Example 1 Interpolation
---

```python
# Import each package
import numpy as np
import scipy.interpolate as interpolate
from matplotlib import pyplot as plt

# Set the figure size
plt.rcParams["figure.figsize"] = (12, 9)

# Set the original data
x = np.linspace(0, 10, 11)
y = np.cos(x ** 2 / 9)

# Define the new x values
x_new = np.linspace(0, 10, 41)

# Perform linear interpolation
linear = interpolate.interp1d(x, y)

# Perform cubic interpolation
cubic = interpolate.interp1d(x, y, 'cubic')

# Plot the results of the interpolation
plt.plot(x, y, 'o', x_new, linear(x_new), '-', x_new, cubic(x_new), '--')
plt.xlabel('X', size=16)
plt.ylabel('Y', size=16)
plt.title('Interpolation Example', size=18)

# Display the legend
plt.legend(['data', 'linear', 'cubic'], loc='best')

# Set the tick label size
plt.tick_params(labelsize=16)

# Show the figure
plt.show()
```
