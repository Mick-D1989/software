## Example 2.12.8 Stripmap NESZ
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light, pi, Boltzmann

from matplotlib import pyplot as plt
```

**Set the range resolution (m)**

```python
range_resolution = [1, 5, 10]
```

**Set the average transmitted power (W)**

```python
average_power = 800
```

**Set the antenna effective aperture (m^2)**

```python
effective_aperture = 3.5
```

**Set the operating frequency (Hz)**

```python
operating_frequency = 900e6
```

**Calculate the wavelength**

```python
wavelength = speed_of_light / operating_frequency
```

**Set the slant range (m)**

```python
slant_range = 200e3
```

**Set the system temperature (k)**

```python
system_temperature = 290
```

**Set the noise figure (linear)**

```python
noise_figure = 4
```

**Set the losses (linear)**

```python
losses = 6
```

**Set the velocity (m/s)**

```python
velocity = np.linspace(10, 250)
```

**Calculate the NESZ and Plot the Results**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Calculate the NESZ

for delta_r in range_resolution:
    
    nesz = (8 * pi * wavelength * slant_range**3 * velocity * system_temperature * Boltzmann * noise_figure * losses) / (average_power * effective_aperture**2 * delta_r)
    
    plt.plot(velocity, 10 * np.log10(nesz), linewidth=2, label=f'$\delta_r$ = {delta_r} m')



# Set the plot title and labels

plt.title('Stripmap Noise Equivalent Sigma Zero', size=18)

plt.xlabel('Platform Velocity (m/s)', size=16)

plt.ylabel('NESZ (dB)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')



# Show the figure

plt.show()
```
