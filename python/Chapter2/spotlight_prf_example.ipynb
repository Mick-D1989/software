## Example 2.12.12 Spotlight SAR Pulse Repetition Frequency
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light

from matplotlib import pyplot as plt
```

**Set the operating frequency (Hz)**

```python
operating_frequency = 12e9
```

**Calculate the wavelength (m)**

```python
wavelength = speed_of_light / operating_frequency
```

**Set the aperture length (m)**

```python
aperture_length = 350
```

**Set the unambiguous range (m)**

```python
unambiguous_range = 250e3;
```

**Set the velocity (m/s)**

```python
velocity = np.linspace(50, 300)
```

**Set the slant range (m)**

```python
slant_range = 20e3
```

**Calculate the upper limit (Hz)**

```python
prf_upper_limit = speed_of_light / (2 * unambiguous_range) * np.ones_like(velocity)
```

**Calculate the lower limit (Hz)**

```python
prf_lower_limit = 2 * velocity * aperture_length / (wavelength * slant_range)
```

**Plot the upper and lower limtis**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Upper and lower limits

plt.plot(velocity, prf_upper_limit, linewidth=2, label='Upper Limit')

plt.plot(velocity, prf_lower_limit, linewidth=2, label='Lower Limit')



# Set the plot title and labels

plt.title('Spotlight PRF Upper and Lower Limits', size=18)

plt.xlabel('Platform Velocity (m/s)', size=16)

plt.ylabel('Pulse Repetition Frequency (Hz)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')



# Display the figure

plt.show()
```
