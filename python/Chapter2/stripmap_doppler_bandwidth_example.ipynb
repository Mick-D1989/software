## Example 2.12.4 Stripmap SAR Doppler Bandwidth
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light

from matplotlib import pyplot as plt
```

**Set the velocity (m/s)**

```python
velocity = np.linspace(50, 300)
```

**Set the operating frequency (Hz)**

```python
operating_frequency = 3e9
```

**Calculate the wavelength (m)**

```python
wavelength = speed_of_light / operating_frequency
```

**Set the antenna beamwidth (deg)**

```python
beamwidths = [1, 2.5, 10]
```

**Calculate the Doppler bandwidth (Hz)**
**And plot the results**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)


for beamwidth in beamwidths:
    
    doppler_bandwidth = 4 * velocity / wavelength * np.sin(np.radians(beamwidth / 2))
    
    plt.plot(velocity, doppler_bandwidth, linewidth=2, label=f'Beamwidth = {beamwidth}$^\circ$')
    
    
# Set the plot title and labels

plt.title('Stripmap Doppler Bandwidth', size=18)

plt.xlabel('Platform Velocity (m/s)', size=16)

plt.ylabel('Doppler Bandwidth (Hz)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')



# Display the figure

plt.show()
```
