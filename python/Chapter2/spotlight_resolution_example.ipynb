## Example 2.12.9 Spotlight Cross-Range Resolution
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light

from matplotlib import pyplot as plt
```

**Set the operating frequency (Hz)**

```python
operating_frequency = [1e9, 2e9, 6e9, 10e9]

labels = ['L-band', 'S-band', 'C-band', 'X-band']
```

**Set the slant range (m)**

```python
slant_range = 10e3
```

**Set the synthetic aperture length (m)**

```python
aperture_length = np.linspace(10, 200, 500)
```

**Calculate the cross-range resolution and plot the results**

```python
# Set the figure size
plt.rcParams["figure.figsize"] = (12, 9)


# Calculate the cross-range resolution (m)

for frequency, legend_string in zip(operating_frequency, labels):
    
    wavelength = speed_of_light / frequency
    
    cross_range_resolution = wavelength * slant_range / (2 * aperture_length)
    
    plt.plot(aperture_length, cross_range_resolution, linewidth=2, label=legend_string)



# Set the plot title and labels

plt.title('Spotlight Cross-Range Resolution', size=18)

plt.xlabel('Synthetic Aperture Length (m)', size=16)

plt.ylabel('Cross-Range Resolution (m)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')



# Display the figure

plt.show()
```
