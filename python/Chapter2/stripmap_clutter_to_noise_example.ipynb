## Example 2.12.7 Stripmap SAR Clutter-to-Noise Ratio
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light, pi, Boltzmann

from matplotlib import pyplot as plt
```

**Set the range resolution (m)**

```python
range_resolution = 25
```

**Set the backscatter coefficient**

```python
backscatter_coefficient = [0.01, 0.05, 0.25]
```

**Set the average transmitted power (W)**

```python
average_power = 500
```

**Set the antenna effective aperture (m^2)**

```python
effective_aperture = 1
```

**Set the operating frequency (Hz)**

```python
operating_frequency = 3e9
```

**Calculate the wavelength**

```python
wavelength = speed_of_light / operating_frequency
```

**Set the slant range (m)**

```python
slant_range = 100e3
```

**Set the system temperature (K)**

```python
system_temperature = 290
```

**Set the noise figure (dB)**

```python
noise_figure = 9
```

**Set the losses (dB)**

```python
losses = 12
```

**Set the velocity (m/s)**

```python
velocity = np.linspace(25, 300)
```

**Calculate the Clutter-to-Noise Ratio and Plot the Results**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Calculate the clutter-to-noise ratio

for sigma in backscatter_coefficient:
    cnr = average_power * effective_aperture**2 * sigma * range_resolution / (8 * pi * wavelength * slant_range**3 * velocity * system_temperature * Boltzmann * noise_figure * losses)
    plt.plot(velocity, 10 * np.log10(cnr), linewidth=2, label=f'$\sigma_0$ = {sigma} $m^2$')



# Set the plot title and labels

plt.title('Stripmap Clutter-to-Noise Ratio', size=18)

plt.xlabel('Platform Velocity (m/s)', size=16)

plt.ylabel('CLutter-to-Noise (dB)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')



# Display the figure

plt.show()
```
