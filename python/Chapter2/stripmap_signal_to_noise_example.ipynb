## Example 2.12.6 Stripmap SAR Signal-to-Noise Ratio
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light, pi, Boltzmann

from matplotlib import pyplot as plt
```

**Set the cross range resolution (m)**

```python
cross_range_resolution = 15
```

**Set the radar cross section (m^2)**

```python
rcs = [0.05, 1, 10]
```

**Set the average transmitted power (W)**

```python
average_power = 100
```

**Set the antenna effective aperture (m^2)**

```python
effective_aperture = 5
```

**Set the operating frequency (Hz)**

```python
operating_frequency = 10e9
```

**Calculate the wavelength (m)**

```python
wavelength = speed_of_light / operating_frequency
```

**Set the slant range (m)**

```python
slant_range = 175e3
```

**Set the system temperature (K)**

```python
system_temperature = 290
```

**Set the noise figure (linear)**

```python
noise_figure = 6
```

**Set the losses (linear)**

```python
losses = 10
```

**Set the velocity (m/s)**

```python
velocity = np.linspace(25, 300)
```

**Calculate the Signal-to-Noise Ratio and Plot the Results**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Calculate the signal-to-noise ratio

for sigma in rcs:
    
    snr = average_power * effective_aperture**2 * sigma / (8 * pi * wavelength * slant_range**3 * cross_range_resolution * velocity * system_temperature * Boltzmann * noise_figure * losses)
    
    plt.plot(velocity, 10 * np.log10(snr), linewidth=2, label=f'$\sigma$ = {sigma} $m^2$')



# Set the plot title and labels

plt.title('Stripmap Signal-to-Noise Ratio', size=18)

plt.xlabel('Platform Velocity (m/s)', size=16)

plt.ylabel('Signal-to-Noise (dB)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')



# Display the figure

plt.show()
```
