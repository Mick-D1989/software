## Example 2.12.11 Spotlight SAR Doppler Bandwidth
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light

from matplotlib import pyplot as plt
```

**Set the velocity (m/s)**

```python
velocity = np.linspace(50, 300)
```

**Set the operating frequency (Hz)**

```python
operating_frequency = 14e9
```

**Calculate the wavelength (m)**

```python
wavelength = speed_of_light / operating_frequency
```

**Set the synthetic aperture length (m)**

```python
aperture_length = 100
```

**Set the range to image center (m)**

```python
slant_ranges = [10e3, 30e3, 50e3]
```

**Calculate the Doppler bandwidth and plot the results**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Calculate the Doppler bandwidth (Hz) 

for slant_range in slant_ranges:
    
    doppler_bandwidth = 2 * velocity * aperture_length / (wavelength * slant_range);
    
    plt.plot(velocity, doppler_bandwidth, linewidth=2, label=f'Slant Range = {slant_range/1e3} km')
    

    
# Set the plot title and labels

plt.title('Spotlight SAR Doppler Bandwidth', size=18)

plt.xlabel('Platform Velocity (m/s)', size=16)

plt.ylabel('Doppler Bandwidth (Hz)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')
plt.show()
```
