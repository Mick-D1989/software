## Example 2.12.1 Stripmap SAR Range Resolution
---

**Import Modules**

```python
%matplotlib inline

import numpy as np

from scipy.constants import speed_of_light

from matplotlib import pyplot as plt
```

**Set the operating bandwidth (Hz)**

```python
bandwidth = np.linspace(1e6, 20e6)
```

**Set the incident angles (deg)**

```python
incident_angles = [30, 40, 50, 60]
```

**Plot the results**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)


# Plot the ground range resolution

plt.figure


# Loop over the incident angles

for incident_angle in incident_angles:

    ground_range_resolution = speed_of_light / (2 * bandwidth * np.cos(np.radians(incident_angle)))
    
    plt.plot(bandwidth/1e6, ground_range_resolution, linewidth=2, label=f'Incident Angle = {incident_angle}$^\circ$')
    

# Set the plot title and labels

plt.title('Stripmap Ground Plane Resolution', size=18)

plt.xlabel('Bandwidth (MHz)', size=16)

plt.ylabel('Ground Plane Resolution (m)', size=16)


# Set the tick label size

plt.tick_params(labelsize=16)


# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)


# Display the legend

plt.legend(loc='best')



# Display the figure

plt.show()
```
