## Example 7.6.4 Rain Attenuation
---

**Import modules**

```python
%matplotlib inline

import numpy as np

from matplotlib import pyplot as plt

import rain   
```

**Set the operating frequency (GHz)**

```python
frequency = np.linspace(1.0, 100.0)
```

**Set the elevation angle (deg)**

```python
elevation_angle = 10.0
```

**Set the polarization tilt angle (deg)**

```python
polarization_tilt_angle = 0
```

**Set the rain rates (mm/hr)**

```python
slight_rain_rate   = 0.5

moderate_rain_rate = 3.0

heavy_rain_rate    = 6.0

very_rain_rate     = 10.0
```

**Calculate the rain attenuation (dB/km)**

```python
slight     = rain.attenuation(frequency, slight_rain_rate, elevation_angle, polarization_tilt_angle)

moderate   = rain.attenuation(frequency, moderate_rain_rate, elevation_angle, polarization_tilt_angle)

heavy      = rain.attenuation(frequency, heavy_rain_rate, elevation_angle, polarization_tilt_angle)

very_heavy = rain.attenuation(frequency, very_rain_rate, elevation_angle, polarization_tilt_angle)
```

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Plot the attenuation for each rain rate
plt.figure

plt.plot(frequency, slight, linewidth=2, label='Slight Rain')

plt.plot(frequency, moderate, linewidth=2, label='Moderate Rain')

plt.plot(frequency, heavy, linewidth=2, label='Heavy Rain')

plt.plot(frequency, very_heavy, linewidth=2, label='Very Heavy Rain')




# Set the plot title and labels

plt.title('Rain Attenuation', size=18)

plt.xlabel('Frequency (GHz)', size=16)

plt.ylabel('Specific Attenuation (dB/km)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')
plt.show()
```
