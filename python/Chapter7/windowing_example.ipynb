## Example 7.6.2 Windowing Functions
---

**Import modules**

```python
%matplotlib inline

import numpy as np

import scipy.signal.windows as windows

from matplotlib import pyplot as plt

import sys
```

**Set the number of samples in the window**

```python
number_of_samples = 51
```

**Get each window**

```python
rectangular = np.ones(number_of_samples)

hamming = windows.hamming(number_of_samples)

hanning = windows.hann(number_of_samples)

blackman = windows.blackman(number_of_samples)

barthann = windows.barthann(number_of_samples)
```

**Display the windows**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Plot each window

rectangular[0] = 0

rectangular[-1] = 0

plt.plot(rectangular, linewidth=2, label='Rectangular')

plt.plot(hamming, linewidth=2, label='Hamming')

plt.plot(hanning, linewidth=2, label='Hanning')

plt.plot(blackman, linewidth=2, label='Blackman')

plt.plot(barthann, linewidth=2, label='BartHann')



# Set the plot title and labels

plt.title('Window Functions', size=18)

plt.xlabel('Sample Number', size=16)

plt.ylabel('Amplitude (Normalized)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='best')

plt.show()
```

**Calculate the frequency response of each window**

```python
spectrum_rectangular = 10 * np.log10(abs(np.fft.fftshift(np.fft.fft(rectangular, 1024)) + sys.float_info.min))

spectrum_hamming = 10 * np.log10(abs(np.fft.fftshift(np.fft.fft(hamming, 1024)) + sys.float_info.min))

spectrum_hanning = 10 * np.log10(abs(np.fft.fftshift(np.fft.fft(hanning, 1024)) + sys.float_info.min))

spectrum_blackman = 10 * np.log10(abs(np.fft.fftshift(np.fft.fft(blackman, 1024)) + sys.float_info.min))

spectrum_barthann = 10 * np.log10(abs(np.fft.fftshift(np.fft.fft(barthann, 1024)) + sys.float_info.min))
```

**Display the spectrum**

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (12, 9)



# Plot spectrum of each window

plt.plot(spectrum_rectangular, linewidth=2, label='Rectangular')

plt.plot(spectrum_hamming, linewidth=2, label='Hamming')

plt.plot(spectrum_hanning, linewidth=2, label='Hanning')

plt.plot(spectrum_blackman, linewidth=2, label='Blackman')

plt.plot(spectrum_barthann, linewidth=2, label='BartHann')



# Set the y-axis limits (dB)

plt.ylim(-40, 20)



# Set the plot title and labels

plt.title('Window Functions', size=18)

plt.xlabel('Sample Number', size=16)

plt.ylabel('Amplitude (Normalized)', size=16)



# Set the tick label size

plt.tick_params(labelsize=16)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)



# Display the legend

plt.legend(loc='upper left')

plt.show()
```
